```python
#inspecting the data
# Display basic information about the dataset
print(data.info())

# Display summary statistics
print(data.describe())
print("Data Type of 'Date' Column:", data['Date'].dtype)

```

```python
# Handle Missing Values:
# Drop rows with missing values
# df = data.dropna()

# # Impute missing values (example using mean)
# df['Column_Name'].fillna(df['Column_Name'].mean(), inplace=True)


# Convert date column to datetime type
data['Date'] = pd.to_datetime(data['Date'])
print(data)
print("Data Type of 'Date' Column:", data['Date'].dtype)
```

```python
# Remove Unnecessary Columns:
# df = df.drop(['Column1', 'Column2'], axis=1)

# Data Transformation:
# Remove dollar signs and convert to numeric
print(data.dtypes)
data[' Manufacturing Price '] = pd.to_numeric(data[' Manufacturing Price '].astype(str).str.replace('$', '', regex=True))
data[' Sale Price '] = pd.to_numeric(data[' Sale Price '].astype(str).str.replace('$', '', regex=True))
print(data[' Manufacturing Price '])
print(data[' Sale Price '])
```
