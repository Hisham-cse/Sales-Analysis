```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("bankig loan.csv")
```

```python
data
```

```python
#Making data set clean to perform numerical operation 
# replaced the $ sign and - to null value(NaN)
data.replace({'\$': ''}, regex=True, inplace=True)
data.replace({'\-':np.nan}, regex=True, inplace=True)
data.replace({'\,':''},regex=True,inplace=True)
data.replace({'\(':''},regex=True,inplace=True)
data.replace({'\)':''},regex=True,inplace=True)
data
```

```python
data.dtypes
```

```python
# Assuming your DataFrame is named 'data'
# Convert specific columns to float
columns_to_convert = [' Manufacturing Price ', ' Sale Price ', ' Gross Sales ', ' Discounts ', '  Sales ', ' COGS ',' Profit ']
data[columns_to_convert] = data[columns_to_convert].astype(float)

# Display the modified DataFrame
data.dtypes
data
```

```python
import pandas as pd

# DataFrame is named 'data'
# Checking for missing values in the entire DataFrame
print("Checking Null values in Entire Dataset",data.isnull())

```

```python
print("Checking Null values in Entire Dataset",data.isnull())
# Checking the total number of missing values in each column
print("Null values in coloumn wise:\n",data.isnull().sum())

# Checking the total number of missing values in the entire DataFrame
print("total no of Null values=",data.isnull().sum().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

column_name = ' Product '  

value_counts = data[column_name].value_counts()

# Your plotting code here
plt.figure(figsize=(6, 6))  
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90,
        colors=plt.cm.Paired.colors)


plt.title(f'Pie Chart for {column_name}')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'data'
# Assuming you want to create a histogram for a specific column, replace 'column_name' with the actual column name
column_name = ' COGS '

# Your plotting code here
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.hist(data[column_name], bins=24, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram for {column_name}')

# Show the grid
plt.grid(axis='y')

# Display the plot
plt.show()
```
